```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ep_2024 = pd.read_excel("C:/Users/mora_/OneDrive/Documentos/Epsilon Engineering/TENDENCIA 2024 Epsilon.xlsx")
print(ep_2024.head(10))
print()
ep_2024.info()
```

```python
ep_2025 = pd.read_excel("C:/Users/mora_/OneDrive/Documentos/Epsilon Engineering/TENDENCIA 2025 Epsilon.xlsx")
print(ep_2025.head(10))
print()
ep_2025.info()
```

```python
ep_2024.plot(kind='bar', x='Mes', title='Ventas por SKU por mes Epsilon 2024', figsize=(20,10))
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
ep_2025.plot(kind='bar', x='Mes', title='Ventas por SKU por mes Epsilon 2025', figsize=(20,10))
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.grid(axis='y')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
# Para 2024
ep_2024_total = ep_2024.drop('TOTAL', errors='ignore')

totales_mensuales = ep_2024_total.select_dtypes(include=['number'])

plt.figure(figsize=(12, 6))
totales_mensuales.plot(kind='bar', color='royalblue', edgecolor='black', figsize=(12, 6))

plt.title('Ventas Totales por Mes (Epsilon 2024)', fontsize=15)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)


# Para 2025
ep_2025_total = ep_2025.drop('TOTAL', errors='ignore')

totales_mensuales = ep_2025_total.select_dtypes(include=['number'])

plt.figure(figsize=(12, 6))
totales_mensuales.plot(kind='bar', color='royalblue', edgecolor='black', figsize=(12, 6))

plt.title('Ventas Totales por Mes (Epsilon 2025)', fontsize=15)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```
